```python
# 필요한 라이브러리 불러오기
import pandas as pd
import numpy as np
from tqdm import tqdm
from tqdm import trange
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen , urlretrieve, Request
from urllib.error   import HTTPError
from urllib.error   import URLError
from selenium import webdriver
from time import sleep, time
from IPython.core.display import clear_output
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import os



def salt (chrome_driver = True) :
    
    # 소금 이미지 크롤링
    chrome_driver = webdriver.Chrome()

    salt_link = []

    for page in range(1, 10) :
        print("*"*50, 'page', page)
        ################
        sleep(2)
        ################
        chrome_driver.get('https://emart.ssg.com/search.ssg?target=all&query=%EC%86%8C%EA%B8%88&page='+str(page))

        sleep(2)

        div = chrome_driver.find_elements(By.CSS_SELECTOR , 'div.tmpl_itemlist div.mnemitem_tx_thmb')
        for li in div :             
            a_tags = li.find_element(By.TAG_NAME, 'a').get_attribute('href')
            salt_link.append(a_tags)

    print(salt_link)

    sleep(10)

    salt_lst = []

    for lnk in salt_link :
        sleep(5)
        try:
            chrome_driver.get(lnk)
            ########   
            sleep(2)
            ########
            span  = chrome_driver.find_element(By.CLASS_NAME, 'cdtl_item_image')
            img   = span.find_element(By.TAG_NAME, 'img').get_attribute('src')
            print(img)
            salt_lst.append(img)
            print('*' * 50)
        except:
            continue

    print(salt_lst)


    img_folder = './소금'

    if not os.path.isdir(img_folder) :
        os.mkdir(img_folder)

    for idx, link in enumerate(salt_lst) :
        try:
            print(idx, '-', link)
            urlretrieve(link, f'./소금/salt_{idx}.png')
        except:
            continue
            
def pepper (chrome_driver = True) :            
    
    # 후추 이미지 크롤링
    chrome_driver = webdriver.Chrome()

    pepper_link = []

    for page in range(1, 7) :
        print("*"*50, 'page', page)
        ################
        sleep(2)
        ################
        chrome_driver.get('https://emart.ssg.com/search.ssg?target=all&query=%ED%9B%84%EC%B6%94&page='+str(page))

        sleep(2)

        div = chrome_driver.find_elements(By.CSS_SELECTOR , 'div.tmpl_itemlist div.mnemitem_tx_thmb')
        for li in div :             
            a_tags = li.find_element(By.TAG_NAME, 'a').get_attribute('href')
            pepper_link.append(a_tags)

    print(pepper_link)

    sleep(10)

    pepper_lst = []

    for lnk in pepper_link :
        sleep(3)
        try:
            chrome_driver.get(lnk)
            ########   
            sleep(2)
            ########
            span  = chrome_driver.find_element(By.CLASS_NAME, 'cdtl_item_image')
            img   = span.find_element(By.TAG_NAME, 'img').get_attribute('src')
            print(img)
            pepper_lst.append(img)
            print('*' * 50)
        except:
            continue

    print(pepper_lst)

    img_folder = './후추'

    if not os.path.isdir(img_folder) :
        os.mkdir(img_folder)

    for idx, link in enumerate(pepper_lst) :
        try:
            print(idx, '-', link)
            urlretrieve(link, f'./후추/pepper_{idx}.png')
        except:
            continue
```
